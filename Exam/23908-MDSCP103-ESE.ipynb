```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
f = lambda x: 4*x[0] + 6*x[1] - 2*x[0]**2 - 2*x[0]*x[1] - 2*x[1]**2
x1_points = np.linspace(-10,10,100)

X,Y = np.meshgrid(x1_points, x1_points)
F = f((X,Y))
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')

ax.set_xlabel('x1', labelpad=10)
ax.set_ylabel('x2', labelpad=30)
ax.set_zlabel('f(X)', labelpad=10)

surf = ax.plot_surface(X,Y,F, cmap='gist_rainbow')
# plt.plot_surface(X,Y,F, cmap='gist_rainbow')
plt.show()
```

```python
df_x1 = lambda x: 4 - 4*x[0] - 2*x[1]
df_x2 = lambda x: 6 - 2*x[0] - 4*x[1]

# let us start by initialising X_0 to 0,0
x_0 = np.array([[2],[3]])

# learning rate
_alpha = 0.01

# iterations
n = 200

x_list = [x_0,]

for i in range(n):
    temp = x_0 + _alpha * np.array([df_x1(x_0), df_x2(x_0)])
    x_list.append(temp)
    x_0 = temp
    if i % (n/10) == 0:
        print(f'Iteration {i}: x1 = {temp[0]}   x2 = {temp[1]}   f(x1,x2) = {f(temp)}')
```
