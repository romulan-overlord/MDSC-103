```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
c1 = lambda x: 4 - 2*x
c2 = lambda x: 2 + 2*x
c3 = lambda x: x - 1

xpoints = np.linspace(0,2,50)

plt.plot(xpoints, [c1(x) for x in xpoints], label='c1')
plt.plot(xpoints, [c2(x) for x in xpoints], label='c2')
plt.plot(xpoints, [c3(x) for x in xpoints], label='c3')
plt.plot(xpoints, [0 for x in xpoints], label='x-axis')
plt.scatter(0.5,3)


plt.legend()
plt.show()
```
