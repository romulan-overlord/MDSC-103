```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import scipy as sp

from scipy.optimize import minimize
```

```python
f = lambda x : ((x[0] - 2)**2) + 4 * (x[1] - 6)**2

x1 = np.linspace(-10,10,100)
x2 = np.linspace(-10,10,100)

X,Y = np.meshgrid(x1,x2)
F = f((X,Y))

fig = plt.figure(figsize=(15,15))
# ax = plt.axes(projection='3d')

# ax.set_xlabel('x1', labelpad=10)
# ax.set_ylabel('x2', labelpad=30)
# ax.set_zlabel('f(X)', labelpad=10)

# surf = ax.plot_surface(X,Y,F, cmap='gist_rainbow')
# # plt.plot_surface(X,Y,F, cmap='gist_rainbow')
# plt.show()

f2 = lambda x2 : -0.5 * (x2 + 1)**2 + 2

xp = np.linspace(0,5,100)

plt.plot(xp, f2(xp))

plt.contour(X, Y, F, 50, cmap='jet')
plt.scatter(0,1,f((0,1)))

plt.plot()

plt.show()
```

```python
cons = ({'type' : 'ineq', 'fun' : lambda x: -6 * x[0] - 3*(x[1] + 1)**2 + 12})

bnds = ((0,None), (0,None))

x0 = (1,1)

res = minimize(f,x0,bounds=bnds,constraints=cons)

res
```
